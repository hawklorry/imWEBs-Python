```python
class PairList:
    def __init__(self):
        self.data = {}

    def contains(self, id1, id2):
        return (id1, id2) in self.data

    def setValue(self, id1, id2, value):
        self.data[(id1, id2)] = value

    def getValue(self, id1, id2):
        return self.data.get((id1, id2), 0)

    def add(self, id1, id2, value):
        self.data[(id1, id2)] = value


def get_overlap_relation_pair_list(spatial1_ras:Raster, spatial2_ras:Raster):
    pl = PairList()

    rows = spatial1_ras.configs.rows
    cols = spatial1_ras.configs.columns

    cell_area = spatial1_ras.configs.resolution_x * spatial1_ras.configs.resolution_y

    for col in range(cols):
        for row in range(rows):
            id1 = int(spatial1_ras[row, col])
            id2 = int(spatial2_ras[row, col])

            if id1 > 0 and id2 > 0:
                if pl.contains(id1, id2):
                    pl.setValue(id1, id2, pl.getValue(id1, id2) + cell_area)
                else:
                    pl.add(id1, id2, cell_area)

    return pl

pl = get_overlap_relation_pair_list(field_raster, subbasin_raster)
print(pl.data)
```
