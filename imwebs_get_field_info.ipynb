```python
import geopandas as gpd
import os
```

```python
field_shp = r"D:\imWEBs\STC\STC-Landuse-Data\field\unit_management_from_raster_dissolved.shp"
output_path = r"D:\imWEBs\STC\STC-Landuse-Data\Processed"

field_shp_df = gpd.read_file(field_shp)
field_shp_df['Area_Ha'] = field_shp_df.geometry.area / 10000
field_shp_df = field_shp_df.rename(columns={"LAND_ID": "ID"})

field_shp_df.to_csv(os.path.join(output_path, 'field_info.csv'), index = False, columns = ['ID','Area_Ha'])
print("done")
```
