using **point_source_shapefile**. One example is given below.

*\[reach_bmp\]*

*point_source_shapefile = PointSource.shp*

## Attributes

-   **Operation** - The operation date in YYYY-MM-DD format
-   **TableName** - The table name in hydroclimate database that defines
    time series of the flow, sediment and nutrients. The table must have
    following columns. The columns are case-insensitive and could be in
    any sequence.
    -   DATE
    -   FLOW
    -   SEDIMENT
    -   ORGANICN
    -   ORGANICP
    -   NO3
    -   NH3
    -   NO2
    -   MINERALP
