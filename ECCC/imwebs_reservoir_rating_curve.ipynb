```python
import pandas as pd
import sqlite3
import os
```

```python
#function to save data frame to sqlite table
def df2sqlite(df, tablename):
    conn = sqlite3.connect(r"D:\imWEBs\STC\model\STC-Catchabasin-Reservoir-Separated\Model01\BMP.db3") 
    c = conn.cursor()

    c.execute(f'CREATE TABLE IF NOT EXISTS {tablename} (storage number, area number, discharge number)')
    conn.commit()

    df.to_sql(tablename, conn, if_exists='replace', index = False)
```

```python
#small dam rating curve folder
rating_curve_path = r"D:\git\imWEBs\IMWEBsInterface\resources\DemoDatasets\STCData\STC_Small_Dam"

for rating_curve_file in os.listdir(rating_curve_path):
    if ".dam" not in rating_curve_file:
        continue

    print(rating_curve_file)

    df = pd.read_csv(os.path.join(rating_curve_path, rating_curve_file), skiprows=3, skip_blank_lines=True, names = ["storage","area","discharge"], delim_whitespace = True)
    df2sqlite(df, os.path.splitext(rating_curve_file)[0])
```
