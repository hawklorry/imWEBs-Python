```python
#existing, wetland loss, wetland restoration

#wetland enhancement 5m
model_path = r"D:\imWEBs\BRC\model\Enhancement_5m\Enhancement_5m"
wetland_shp = "D:/git/imWEBs/IMWEBsInterface/resources/DemoDatasets/BRC2021/SubArea_Enhancement_5m/wetland_enhancement_5m.shp"
wetland_csv = "D:/git/imWEBs/IMWEBsInterface/resources/DemoDatasets/BRC2021/SubArea_Enhancement_5m/wetland_enhancement_5m.csv"

#wetland enhancement 15m
#model_path = r"D:\imWEBs\BRC\model\Enhancement_15m\Enhancement_15m"
#wetland_shp = "D:/git/imWEBs/IMWEBsInterface/resources/DemoDatasets/BRC2021/SubArea_Enhancement_15m/wetland_enhancement_15m.shp"

#wetland enhancement 30m
#model_path = r"D:\imWEBs\BRC\model\Enhancement_30m\Enhancement_30m"
#wetland_shp = "D:/git/imWEBs/IMWEBsInterface/resources/DemoDatasets/BRC2021/SubArea_Enhancement_30m/wetland_enhancement_30m.shp"

wetland_type_col = "IMPACT"
wetland_id_lookup_csv = r"D:\imWEBs\BRC\wetland_id_lookup.csv"
wetland_type_list = ['Drained/Altered', 'Drained/Consolidated']

#don't change from here
#---------------------------------------------------------------------
normal_surface_area_col = "NormalArea_ha"
normal_storage_col = "NormalVolume_104M3"
max_surface_area_col = "MaxArea_ha"
max_storage_col = "MaxVolume_104M3"
```

```python
def storage_from_area(area_ha):
    if area_ha <= 70:
        return 2.85 * area_ha**1.22 / 10.0
    else:
        return 7.1 * area_ha + 9.97 / 10.0
```

```python
# add CellIndex, SubAreaId to export file CellSubarea.csv
wetland_shp_df = gpd.read_file(wetland_shp)

#max
wetland_shp_df[max_surface_area_col] = wetland_shp_df.area / 10000
wetland_shp_df[max_storage_col] = wetland_shp_df[max_surface_area_col].apply(storage_from_area)

#normal
wetland_shp_df[normal_surface_area_col] = np.where(wetland_shp_df[wetland_type_col] == 'Altered', wetland_shp_df[max_surface_area_col]/3.0, wetland_shp_df[max_surface_area_col])
wetland_shp_df[normal_storage_col] = wetland_shp_df[normal_surface_area_col].apply(storage_from_area)

#save back to shapefile
wetland_shp_df.to_file(wetland_shp)

#link to imwebs id
wetland_id_lookup_csv = pd.read_csv(wetland_id_lookup_csv)
wetland_id_lookup_csv['OBJECTID'] =wetland_id_lookup_csv['OriginalID']
wetland_df = wetland_id_lookup_csv.merge(wetland_shp_df, on = 'OBJECTID', how = 'inner')


#select by type
wetland_selected1 = wetland_df[wetland_df[wetland_type_col].str.contains('Drained/Altered')]
wetland_selected2 = wetland_df[wetland_df[wetland_type_col] == 'Drained/Consolidated']
#wetland_df = pd.concat([wetland_selected1, wetland_selected2])
#wetland_df.to_csv(wetland_csv, index = False, columns= ['imWEBsID',wetland_type_col, normal_surface_area_col,normal_storage_col, max_surface_area_col,max_storage_col])

```
