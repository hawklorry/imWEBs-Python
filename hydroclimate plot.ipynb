```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect(r"D:\imWEBs\STC\hydroclimate\HydroClimate_STC.db3")
```

```python
#get all table names
query = "select TABLENAME from stations"
df_tables = pd.read_sql_query(query,conn)

for table in df_tables.TABLENAME:
    if "PCP" not in table: 
        continue

    query = "SELECT DATES, VALUE FROM " + table

    df_data = pd.read_sql_query(query,conn, parse_dates = ["DATES"], index_col="DATES")
    
    #plot each year
    for year in range(2000, 2018):
        print(f"{table} - {year} - {df_data.loc[str(year)].VALUE.sum()}")
        
```

```python
#get all table names
query = "select TABLENAME from stations"
df_tables = pd.read_sql_query(query,conn)

for table in df_tables.TABLENAME:
    query = "SELECT DATES, VALUE FROM " + table

    df_data = pd.read_sql_query(query,conn, parse_dates = ["DATES"], index_col="DATES")
    
    #plot each year
    for year in range(2000, 2018):
        df_data.loc[str(year)].plot(title = f"{table} - {year}")
        plt.show()
```
