```python
import geopandas as gpd
import numpy as np
import pandas as pd
```

```python
shp_path = r"D:\imWEBs\STC\STC-Landuse-Data\All Years\shapefile"
output_csv_file = r"D:\imWEBs\STC\STC-Landuse-Data\All Years\crop_codes.csv"
```

```python
years = np.arange(start = 1991, stop = 2021, step = 1)
years
```

```python
#add crop colume if it's not in the shapefile
for year in years:
    print(year)
    shp = gpd.read_file(f"{shp_path}/land{year}_by_land_id.shp")
    if 'CROP' not in shp.columns:
        if 'CROPCODE' in shp.columns:
            shp['CROP'] = shp['CROPCODE']
            print(f"{year}-'CROPCODE'")
        if 'CROP_CODE' in shp.columns:
            shp['CROP'] = shp['CROP_CODE']
            print(f"{year}-'CROP_CODE'")
        if 'DN' in shp.columns:
            shp['CROP'] = shp['DN']
            print(f"{year}-'DN'")
        #shp.to_file(f"{shp_path}/land{year}_by_land_id.shp")
```

```python
crops = {}

for year in years:
    print(year)
    shp = gpd.read_file(f"{shp_path}/land{year}_by_land_id.shp")
    crops[year] = shp['CROP'].unique()
    crops[year].sort()
crops
```

```python
df = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in crops.items() ]))
df
```

```python
df.to_csv(output_csv_file)
```
