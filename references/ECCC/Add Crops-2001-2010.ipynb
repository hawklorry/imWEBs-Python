```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
crop_excel = r"D:\imWEBs\STC\STC-Landuse-Data\STC-landmanagement-2001-2010\STC Dataset 1991-2010 Complete-Names Removed.xlsx"
```

```python
crops = pd.read_excel(crop_excel)
```

```python
shp_path = r"D:\imWEBs\STC\STC-Landuse-Data\STC-landmanagement-2001-2010"
out_path = r"D:\imWEBs\STC\STC-Landuse-Data\STC-landmanagement-2001-2010\add_crop"
```

```python
years = np.arange(start = 2001, stop = 2011, step = 1)
years
```

```python
crops.columns
```

```python
crops = crops[['Land_id','Year','Crop']]
```

```python
crops = crops.rename(columns={"Land_id":"LAND_ID", "Crop": "CROP"})
crops
```

```python
for year in years:
    print(year)
    shp = gpd.read_file(f"{shp_path}/land{year}_by_land_id.shp")    
    shp = shp.merge(crops[crops['Year'] == year], how='left', on='LAND_ID')    
    shp.drop(columns = 'Year')
    shp.to_file(f"{out_path}/land{year}_by_land_id.shp")
    
```
