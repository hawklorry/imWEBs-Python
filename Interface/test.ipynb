```python
def find_bmp(tablename:str):
    bmp = bmp_names_df[bmp_names_df["ParameterTable"].str.lower() == tablename.lower()]
    if len(bmp) > 0:
        bmp_id = bmp.index[-1]
        bmp = bmp.loc[bmp.index[-1]]
        return f'\t"""Parameter Table for BMP: {bmp.Name} ({bmp_id})"""'
    
    bmp = bmp_names_df[bmp_names_df["ManagementTable"].str.lower() == tablename.lower()]
    if len(bmp) > 0:
        bmp_id = bmp.index[-1]
        bmp = bmp.loc[bmp.index[-1]]
        return f'\t"""Distribution Table for BMP: {bmp.Name} ({bmp_id})"""'
    
    bmp = bmp_names_df[bmp_names_df["AllItemTable"].str.lower() == tablename.lower()]
    if len(bmp) > 0:
        bmp_id = bmp.index[-1]
        bmp = bmp.loc[bmp.index[-1]]
        return f'\t"""Distribution Table for BMP: {bmp.Name} ({bmp_id})"""'
    
    return None
        
        
```

```python
bmp_def_df = pd.read_csv(r"C:\Work\imWEBs\Interface\bmp_db_structure.csv")
bmp_def_df.head()
```

```python
group = bmp_def_df.groupby("TableName")
group
```

```python
for g in group:
    print("\n\n")
    table_name = g[0]
    table_def = g[1]
    print(f"class {table_name}(BMPTable):")    

    count = 0
    for index in table_def.index:
        if count == 0:
            table_name = table_def.loc[index, "TableName"]
            class_description = find_bmp(table_name)
            if class_description is not None:
                print(class_description)
            print(f"\t__tablename__ = '{table_name}'")
        count = count + 1


        description = table_def.loc[index, "Description"]
        if len(str(description)) > 1:
            print(f'\t"""{description}"""')

        name = table_def.loc[index, "ColumnName"]
        type = table_def.loc[index, "ColumnType"]

        if str(type).lower() == "int":
            type = "Integer"
        elif str(type).lower() == "text":
            type = "String"
        elif str(type).lower() == "real":
            type = "Float"
        elif str(type).lower() == "double":
            type = "Double"
        
        print(f"\t{name} = Column({type})")
```
